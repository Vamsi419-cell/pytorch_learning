# imports


```python
import torch
import numpy as np
import pandas as pd
```

# how do we create a "tensor"


```python
#using python list
x = torch.tensor([1,2,3],dtype= torch.float32)
print("Tensor data type",x.dtype)
```

```python
#from numpy array
numpy_array = np.array([[1,2,3],[4,5,6]])
tensor = torch.from_numpy(numpy_array)
print("Tensor from numpy: \n\n",tensor)
```

```python
# All zeros , all ones and rand
zeros = torch.zeros(2,3)
print(f"Tensor with zeros \n{zeros}")
ones = torch.ones(2,3)
print(f"Tensor with ones \n {ones}")
rand = torch.rand(2,3)
print(f"Tensor with randoms \n {rand}")
range = torch.arange(0,10,step =1)
print(f"Arange {range}")
```
